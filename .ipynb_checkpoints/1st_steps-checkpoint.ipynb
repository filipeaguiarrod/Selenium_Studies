## Pegar todos os links de aulas

- Pegar todos os links da aula;
    - Criar dic{'Nome da Aula':link}
- Navegar até exercício 3;
    - Achar link da url 3 e acessar.

```python
browser = Chrome() #Abre o navegador

browser.get('https://selenium.dunossauro.live/aula_04.html') # Acessa link específico no navegador;
```

```python
U_lista = browser.find_element_by_tag_name('aside') # Aside dentro do html
Lista_aulas = U_lista.find_elements_by_tag_name('a') # Listas de aula que esta dentro do último âncoras (a)
```

```python
Lista_links = [i.get_attribute('href') for i in Lista_aulas]
Lista_links
```

```python
Lista_nome = [i.text for i in Lista_aulas]
Lista_nome
```

```python
# Criando Dicionário com links e nomes

Aurelio={}

for i in Lista_aulas:
    
    Aurelio[i.text] = i.get_attribute('href')
```

```python
browser.get(Aurelio['Aula 3'])
```

```python
#### Criando função de pegar os links e fazer dicionário:

def get_links(link, elemento):
        
    browser = Chrome() #Abre o navegador

    browser.get(link)
    
    ''' Dado um Site !
    Pega todos links dentro de um elemento
    - browser = instância do navegador Chorme
    - elemento = web_element'''
        
    Aurelio={}
    
    sleep(3)
    
    Elemento = browser.find_element_by_tag_name(elemento) # Aside dentro do html
    
    Lista_elementos = Elemento.find_elements_by_tag_name('a') # Listas de aula que esta dentro do último âncoras (a)
    
    for i in Lista_elementos:
    
        Aurelio[i.text] = i.get_attribute('href')
        
    return Aurelio    
```

```python
get_links('https://selenium.dunossauro.live/aula_04.html','main')
```
